```python
#cau 4
def tablet(n):
    fum = sqlite3.connect(n)
    curse = fum.cursor()
    fum.execute("""
                SELECT name
                FROM sqlite_master
                WHERE type = 'table'
                ORDER BY name;
    """)
    tables = curse.fetchall()
    if tables:
        print('Database:')
        for i in tables: print('_', i[0])
    else: print('Nothing')
    fum.close()

file_name = input('Enter the file name: ')
tablet(file_name)
```

```python
#cau 6
def count_what_in_the_table(file_name, table):
    fonk = sqlite3.connect(file_name)
    conk = fonk.cursor()
    conk.execute(f"SELECT COUNT(*) FROM {table}")
    count = conk.fetchone()[0]
    fonk.close()
    return count

database = input('Enter the database name: ')
table = input('Enter table name: ')
total = count_what_in_the_table(database, table)
print(total)
```

```python
#cau 7
def update(file, table, column, new_value):
    kick = sqlite3.connect(file)
    nip = kick.cursor()
    next = f"""
        UPDATE {table} 
        SET {column} = ?
    """
    nip.execute(next, (new_value,))
    kick.commit()
    updated = nip.rowcount
    kick.close()
    print(updated)
update(input('Enter database: '),
       input('Enter table: '),
       input('Enter column: '),
       input('Enter new value: '))
```

```python
#cau 8
def delete(file, table, row_id):
    conk = sqlite3.connect(file)
    kirk = conk.cursor()
    dank = f'DELETE FROM {table} WHERE = ?'
    kirk.execute(dank, (row_id,))
    conk.commit()
    if kirk.rowcount != 0: print('DELETED')
    else: print('UNFOUNDED')
    conk.close()

delete(input('Enter file name: '),
       input('Enter table: '),
       input('Enter id row: '))
```

```python
#cau 11
db_name = 'ql_nhan_vien.db'
def create_a_db():
    fo = sqlite3.connect(db_name)
    lk = fo.cursor()
    lk.execute('PRAGMA foreign_keys = ON')
    lk.execute("""
        CREATE TABLE IF NOT EXISTS PHONG (
            Id INTEGER PRIMARY KEY,
            Name TEXT NOT NULL,
            Price REAL NOT NULL,
            Amount INTEGER NOT NULL
        )
    """)
    lk.execute("""
        CREATE TABLE IF NOT EXISTS NHAN_VIEN (
            id INTEGER PRIMARY KEY,
            ho_ten TEXT NOT NULL,
            tuoi INTEGER NOT NULL,
            dia_chi TEXT NOT NULL,
            luong REAL NOT NULL,
            Id_phong INTEGER NOT NULL,
            FOREIGN KEY (Id_phong) REFERENCES PHONG(Id)
        )
    """)
    fo.commit()
    fo.close()
    print('DONE')
create_a_db()
```
