```python
#bai 5
mlb = np.genfromtxt('mlb_height_weight.txt', delimiter= ',', dtype= None, encoding= None)
rows = np.array([row.split('\t') for row in mlb])
rows = rows[1:]
dtype = [
    ('Name', 'U50'),
    ('Team', 'U10'),
    ('Position', 'U20'),
    ('Height', 'f4'),
    ('Weight', 'f4'),
    ('Age', 'f4')
]
structured = np.zeros(rows.shape[0], dtype=dtype)

structured['Name'] = rows[:, 0]
structured['Team'] = rows[:, 1]
structured['Position'] = rows[:, 2]
structured['Height'] = rows[:, 3].astype(float)
structured['Weight'] = rows[:, 4].astype(float)
structured['Age'] = rows[:, 5].astype(float)
#1
arr_height = structured['Height']
print(arr_height)
#2
arr_weight = structured['Weight']
print(arr_weight)
#3
arr_height_m = arr_height * 0.0254
print(arr_height_m)
#4
arr_weight_m = arr_weight * 0.453592
print(arr_weight_m)
#5
arr_bmi = arr_weight_m/(arr_height_m**2)
print(arr_bmi)
#6
arr_weight_kg = structured['Weight'][50]
print(arr_weight_kg)
#7
arr_height_m_100 = structured['Height'][100: 110]
print(arr_height_m_100)
#8
print(arr_bmi[arr_bmi < 21])
#9
print(arr_height.mean())
print(arr_weight.mean())
#10
print(arr_height.max())
print(arr_weight.max())
#11
print(arr_height.min())
print(arr_weight.min())
```
