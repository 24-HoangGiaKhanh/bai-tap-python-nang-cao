```python
#cau 1
import threading
def name(): print(f'Thread name: ', threading.current_thread().name)

threads = []

for i in range(5):
    t = threading.Thread(target= name, name= f'Thread-{i+1}')
    threads.append(t)
    t.start()

for i in threads: i.join()
```

```python
#cau 2
def download_file(file_name): print(f"Downloading: {file_name} ({threading.current_thread().name})")

files = ["file1.txt", "file2.txt", "file3.txt"]

threads = []

for f in files:
    t = threading.Thread(target=download_file, args=(f,))
    threads.append(t)
    t.start()

for t in threads: t.join()
```

```python
#cau 3
def even():
    for i in range(30, 51):
        if i % 2 == 0: print(f'This is even: {i}')

def odd():
    for i in range(30, 51):
        if i % 2 != 0: print(f'This is odd: {i}')

t1 = threading.Thread(target= even, name= 'Even threads')
t2 = threading.Thread(target= odd, name= 'Odd threads')

t1.start()
t2.start()

t1.join()
t2.join()
```

```python
#cau 4
partial = []

lock = threading.Lock()

def factorial_part(x, y):
    result = 1
    for i in range(x, y + 1): result *= i

    with lock: partial.append(result)

def factorial_multithread(n, num_threads= 2):
    threads = []
    step = n // num_threads

    start = 1
    for i in range(num_threads):
        end = n if i == num_threads - 1 else start + step - 1
        t = threading.Thread(target=factorial_part, args= (start, end))
        threads.append(t)
        t.start()
        start = end + 1

    for t in threads: t.join()

    result = 1
    for value in partial: result *= value

    return result
```

```python
#cau 5
import requests

def process_http_request(n):
    try:
        print(f"Sending request to {n} ({threading.current_thread().name})")
        response = requests.get(n)
        print(f"{n}: Status {response.status_code}")
    except Exception as e: print(f"Error with {n}: {e}")

urls = ["https://www.youtube.com/watch?v=dQw4w9WgXcQ", 
        "https://youtu.be/FKwVFWwA_x8?list=RDFKwVFWwA_x8", 
        "https://youtu.be/N4lQtxmOwSg?list=RDN4lQtxmOwSg",]

threads = []

for i in urls:
    t = threading.Thread(target= process_http_request, args= (i, ))
    threads.append(t)
    t.start()

for t in threads: t.join()
```

```python
#cau 6
WHAT = 100

def print_even():
    for num in range(0, WHAT + 1, 2): print(f"Even: {num}")

def print_odd():
    for num in range(1, WHAT + 1, 2): print(f"Odd: {num}")

even_thread = threading.Thread(target=print_even)
odd_thread = threading.Thread(target=print_odd)

even_thread.start()
odd_thread.start()

even_thread.join()
odd_thread.join()
```

```python
#cau 7
consoles = [
    "Starting Main Thread",
    "Starting Google",
    "Starting Yahoo",
    "Starting Facebook",
    "Google: Web Aug 05 15:20:55 2023",
    "Yahoo: Web Aug 05 15:20:56 2023",
    "Google: Web Aug 05 15:20:56 2023",
    "Facebook: Web Aug 05 15:20:57 2023",
    "Google: Web Aug 05 15:20:57 2023",
    "Yahoo: Web Aug 05 15:20:58 2023",
    "Google: Web Aug 05 15:20:58 2023",
    "Exiting Google",
    "Yahoo: Web Aug 05 15:21:00 2023",
    "Facebook: Web Aug 05 15:21:00 2023",
    "Yahoo: Web Aug 05 15:21:05 2023",
    "Exiting Yahoo",
    "Facebook: Web Aug 05 15:21:03 2023",
    "Facebook: Web Aug 05 15:21:06 2023",
    "Exiting Facebook",
    "Exiting Main Thread"
]

f = [threading.Event() for _ in consoles]

def print_line(index):
    f[index].wait()       
    print(consoles[index])
    if index + 1 < len(f): f[index + 1].set()  

threads = [threading.Thread(target=print_line, args=(i, )) for i in range(1, len(consoles))]

for t in threads: t.start()

f[0].set()
print(consoles[0])  
f[1].set()     

for t in threads: t.join()
```

```python
#cau 8
import random

n = int(input('Enter the number of element: '))  
main_list = [random.randint(0, 10) for _ in range(n)]
num_threads = int(input('Enter number of thread: ')) 

print("Enter number of elements:", len(main_list))
print("List:", main_list)
print("Enter number of threads:", num_threads)

def print_sublist(start_index, end_index, thread_index):
    sub_list = main_list[start_index:end_index]
    print(f"Thread {thread_index + 1}:", sub_list)

size = len(main_list) // num_threads
threads = []

for i in range(num_threads):
    start = i * size
    end = (i + 1) * size if i != num_threads - 1 else len(main_list)
    t = threading.Thread(target=print_sublist, args=(start, end, i))
    threads.append(t)
    t.start()

for t in threads: t.join()

total = sum(main_list)
print('total list:', total)
```

```python
#cau 10
import random

n = n = int(input('Enter the number of element: '))   
main_list = [random.randint(0, 100) for _ in range(n)]
num_threads = int(input('Enter number of thread: ')) 

print('Enter the number of element: ', len(main_list))
print('List:', main_list)
print('Enter number of threads:', num_threads)

def find_sublist_max(start, end, thread_index):
    sublist = main_list[start: end]
    print(f'Thread {thread_index + 1}:', sublist)

size = len(main_list) // num_threads
threads = []

for i in range(num_threads):
    start = i * size
    end = (i + 1) * size if i != num_threads - 1 else len(main_list)
    t = threading.Thread(target= find_sublist_max, args= (start, end, i))
    threads.append(t)
    t.start()

for t in threads: t.join()

highest = max(main_list)
print('The maximum number in the list:', highest)
```
