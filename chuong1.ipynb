```python
#bai 1
class rectangle:
    def __init__(self, x: float = 0, y: float = 0):
        self.x = x
        self.y = y

    def display(self):
        while True:
            x = float(input('Nhap chieu dai: '))
            y = float(input('Nhap chieu rong: '))
            if x == y:
                print('Day la hinh vuong!')
                continue
            else:
                text: str = 'Hinh chu nhat'
                print(f'{text:_^100}')
                if x > y: print(f'Chieu dai = {x}\nChieu rong = {y}')
                else: print(f'Chieu dai = {y}\nChieu rong = {x}')
                print(f'Chu vi = {(x + y) * 2:.2f}')
                print(f'Dien tich = {x * y:.2f}')
                break

a = rectangle()
a.display()
```

```python
#bai 2
class Ts:
    def __init__(self, hoten: str = '', toan: float = 0, ly: float = 0, hoa: float = 0):
        self.hoten = hoten
        self.toan = toan
        self.ly = ly
        self.hoa = hoa

    def insert(self):
        self.hoten = input('Nhap ho ten thi sinh: ')
        self.toan = float(input('Nhap diem toan: '))
        self.ly = float(input('Nhap diem ly: '))
        self.hoa = float(input('Nhap diem hoa: '))

    def total(self): return self.toan + self.ly + self.hoa

    def display(self): print(f'Thong tin sinh vien: \nHo ten thi sinh: {self.hoten}\nDiem toan: {self.toan}\nDiem ly: {self.ly}\nDiem hoa: {self.hoa}\nTong diem: {self.total()}')

def Main():
    ds = []
    lits = int(input('Nhap so luong thi sinh: '))
    for i in range(0, lits + 1):
        bs = Ts()
        bs.insert()
        ds.append(bs)
        dat = [bs for _ in ds if bs.total() >= 20]
        dat.sort(key= lambda x: x.total(), reverse= True)
        print('Danh sach trung tuyen:')
        if len(dat) == 0: print('Khong co ai trung tuyen!')
        else:
            print(f'Co {len(dat)} thi sinh trung tuyen!')
            for k in dat: k.display()

if __name__ == '__main__': Main()
```

```python
#bai 3
import math
class phan_so:
    def __init__(self, tu: int = 0, mau: int = 0):
        self.tu = tu
        self.mau = mau

    def insert(self):
        self.tu = int(input('Nhap tu so: '))
        while True:
            self.mau = int(input('Nhap mau so: '))
            if self.mau == 0:
                print('Phan so khong hop le')
                continue
            else:
                ucln = math.gcd(self.tu, self.mau)
                self.tu //= ucln
                self.mau //= ucln
                if self.mau < 0: print(f'Phan so: {-self.tu}/{-self.mau}')
                elif self.mau == 1: print(f'Phan so: {self.tu}')
                elif self.mau == -1: print(f'Phan so: {-self.tu}')
                else: print(f'Phan so: {self.tu}/{self.mau}')
                break

a = phan_so()
a.insert()
```

```python
#bai 7
class Date:
    def __init__(self, day = 1, month = 1, year = 2000):
        self.day = day
        self.month = month
        self.year = year
        self.days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    def display(self): print(f'{self.day:02d}/{self.month:02d}/{self.year}')

    def nam_thuan(self, year):
        if (year % 4 == 0 and year % 100 != 0) and (year % 400 == 0): return True
        return False

    def next(self):
        if self.nam_thuan(self.year): self.days_in_month[1] = 29
        else: self.days_in_month[1] = 28

        if self.day < self.days_in_month[self.month - 1]: self.day += 1
        else:
            self.day = 1
            if self.month == 12:
                self.month = 1
                self.year += 1
            else: self.month += 1

day = int(input('Nhap ngay: '))
month = int(input('Nhap thang: '))
nam = int(input('Nhap nam: '))
date = Date(day, month, nam)
print('Hom nay la ngay: ', date.display())
date.next()
print('Ngay mai la: ', date.display())
```

```python
#bai 9
class DaGiac:
    def __init__(self, *canh): self.canh = canh

    def chu_vi(self): return sum(self.canh)

class TuGiac(DaGiac):
    def __init__(self, a, b, c, d): super().__init__(a, b, c, d)

    def dien_tich(self): return "Cần thêm thông tin về chiều cao hoặc góc để tính diện tích."

class HinhBinhHanh(TuGiac):
    def __init__(self, a, b, h):
        super().__init__(a, b, a, b)
        self.h = h

    def dien_tich(self): return self.canh[0] * self.h

    def __str__(self): return f"Hình bình hành với các cạnh {self.canh[0]} và chiều cao {self.h}"

class HinhChuNhat(HinhBinhHanh):
    def __init__(self, dai, rong): super().__init__(dai, rong, rong)

    def dien_tich(self): return self.canh[0] * self.canh[1]

    def __str__(self): return f"Hình chữ nhật có chiều dài {self.canh[0]} và chiều rộng {self.canh[1]}"

class HinhVuong(HinhChuNhat):
    def __init__(self, canh): super().__init__(canh, canh)

    def dien_tich(self): return self.canh[0] ** 2

    def __str__(self): return f"Hình vuông với mỗi cạnh dài {self.canh[0]}"

a = float(input("Nhập độ dài cạnh đáy hình bình hành: "))
b = float(input("Nhập độ dài cạnh bên hình bình hành: "))
h = float(input("Nhập chiều cao hình bình hành: "))

hbh = HinhBinhHanh(a, b, h)
print("Thông tin hình bình hành:\n", hbh, "\nChu vi:", hbh.chu_vi(), "\nDiện tích:", hbh.dien_tich())

dai = float(input("Nhập chiều dài hình chữ nhật: "))
rong = float(input("Nhập chiều rộng hình chữ nhật: "))
hcn = HinhChuNhat(dai, rong)
print("\nThông tin hình chữ nhật:\n", hcn, "\nChu vi:", hcn.chu_vi(), "\nDiện tích:", hcn.dien_tich())

canh = float(input("Nhập độ dài cạnh hình vuông: "))
hv = HinhVuong(canh)
print("\nThông tin hình vuông:\n", hv, "\nChu vi:", hv.chu_vi(), "\nDiện tích:", hv.dien_tich())
```

```python
#bai 10
import math
class Diem:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    def __str__(self): print(f'{self.x, self.y}')

class Elip(Diem):
    def __init__(self, x, y, a, b):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def dien_tich(self): return math.pi * self.a * self.b

    def __str__(self): return f'Elip co toa tam o, truc lon: {self.a}, truc no: {self.b}'

class DuongTron(Elip):
    def __init__(self, x, y, r):
        super().__init__(x, y, r, r)
        self.r = r

    def dien_tich(self): return math.pi * (self.r)**2

    def __str__(self): return f'Duong tron có toa do tam o, ban kinh: {self.r}'

x = float(input('Nhap hoanh do: '))
y = float(input('Nhap tung do: '))
a = float(input('Nhap truc lon: '))
b = float(input('Nhap truc nho: '))

elip = Elip(x, y, a, b)
print('Elip:\n', elip, '\nDien tich: ', elip.dien_tich())

r = float(input('Nhap ban kinh duong tron: '))
circle = DuongTron(x, y, r)
print('Duong tron\n', circle, '\nDien tich: ', circle.dien_tich())
```

```python
#bai 11
import math
class triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def chu_vi(self): return round(self.a + self.b + self.c, 2)

    def dien_tich(self):
        p = self.chu_vi()/2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def isTriangle(self): return ((self.a + self.b) > self.c) and ((self.a + self.c) > self.b) and ((self.b + self.c) > self.a)

class TamGiacVuong(triangle):
    def __init__(self, a, b):
        self.a = a
        self.b = b
        self.c = math.sqrt(a**2 + b**2)

    def dien_tich(self): return 0.5 * self.a * self.b

class TamGiacCan(triangle):
    def __init__(self, a, b):
        self.a = a
        self.b = a
        self.c = b

    def isTriangle(self): return ((self.a + self.b) > self.c) and ((self.a + self.c) > self.b) and ((self.b + self.c) > self.a)

class TamGiacDeu(triangle):
    def __init__(self, a): self.a = self.b = self.c = a

    def dien_tich(self): return round(math.sqrt(3)/4 * (self.a)**2, 2)

x = TamGiacVuong(4, 5)
print(f'Tam giac vuong: \nChu vi = {x.chu_vi()}\nDien tich = {x.dien_tich()}')

y = TamGiacCan(4, 8)
print(f'Tam giac can: \nChu vi = {y.chu_vi()}\nDien tich = {y.dien_tich()}')

z = TamGiacDeu(6)
print(f'Tam giac deu: \nChu vi = {z.chu_vi()}\nDien tich = {z.dien_tich()}')
```
