```python
#cau 4
from tkinter import messagebox

import tkinter as tk
from tkinter import messagebox

class Student:
    def __init__(self, root):
        self.root = root
        self.root.title("Student Info")
        self.root.geometry("300x200")

        tk.Label(root, text="Name").grid(row= 0, 
                                          column= 0, 
                                          padx= 10, 
                                          pady= 10, 
                                          sticky= 'e')
        self.entry_name = tk.Entry(root)
        self.entry_name.grid(row= 0, 
                             column= 1, 
                             padx= 10, 
                             pady= 10)

        tk.Label(root, text="ID").grid(row= 1, 
                                        column= 0, 
                                        padx= 10, 
                                        pady= 10, 
                                        sticky= 'e')
        self.entry_id = tk.Entry(root)
        self.entry_id.grid(row=1, 
                           column=1, 
                           padx=10, 
                           pady=10)

        tk.Label(root, text="Password").grid(row= 2, 
                                              column= 0, 
                                              padx= 10, 
                                              pady= 10, 
                                              sticky= 'e')
        self.entry_password = tk.Entry(root, show="*")
        self.entry_password.grid(row= 2, 
                                 column= 1, 
                                 padx= 10, 
                                 pady= 10)

        self.send_button = tk.Button(root, 
                                     text="SEND", 
                                     command=self.send_data)
        self.send_button.grid(row=3, 
                              column=0, 
                              columnspan=2, 
                              pady=15)

    def send_data(self):
        messagebox.showinfo("Student Info", f"Name: {self.entry_name.get()}\nID: {self.entry_id.get()}\nPassword: {self.entry_password.get()}")

if __name__ == "__main__":
    root = tk.Tk()
    app = Student(root)
    root.mainloop()
```

```python
#cau 5
import urllib.request

class ImageApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Image")
        self.root.geometry("400x400")

        self.label = tk.Label(root)
        self.label.pack(pady=10)

        tk.Button(root, text="Image", command=self.load).pack(pady=5)
        tk.Button(root, text="Download Image", command=self.load_url).pack(pady=5)

    def load(self):
        self.img = tk.PhotoImage(file="otto.gif")
        self.label.config(image=self.img)

    def load_url(self):
        url = "https://jollycontrarian.com/images/6/6c/Rickroll.jpg"
        filename = "otto.gif"

        urllib.request.urlretrieve(url, filename)

        self.img = tk.PhotoImage(file=filename)
        self.label.config(image=self.img)

if __name__ == "__main__":
    root = tk.Tk()
    app = ImageApp(root)
    root.mainloop()
```

```python
#cau 9
from tkinter import ttk

root = tk.Tk()
root.title('tk')
root.geometry('400x250')

tk.Label(root, 
         text= 'Enter value of m: ').grid(row= 0, 
                                          column= 0, 
                                          padx= 10, 
                                          pady= 10, 
                                          sticky= 'e')
entry_a = tk.Entry(root)
entry_a.grid(row= 0, 
             column= 1, 
             padx= 10, 
             pady= 10)

tk.Label(root, 
         text= 'Enter value of n: ').grid(row= 1, 
                                          column= 0, 
                                          padx= 10, 
                                          pady= 10, 
                                          sticky= 'e')
entry_b = tk.Entry(root)
entry_b.grid(row= 1, 
             column= 1, 
             padx= 10, 
             pady= 10)

tk.Label(root, text= 'Choose function:').grid(row= 2, 
                                              column= 0, 
                                              padx= 10, 
                                              pady= 10, 
                                              sticky= 'e')
combo = ttk.Combobox(root, 
                     values=['GCD', 
                             'LCM'])
combo.current(0) 
combo.grid(row= 2, 
           column= 1, 
           padx= 10, 
           pady= 10)

result_label = tk.Label(root, 
                        text= ' ', 
                        font=('Arial', 10))
result_label.grid(row= 3, 
                  column= 0, 
                  columnspan= 3, 
                  pady= 20)

def calculate():
    a = int(entry_a.get())
    b = int(entry_b.get())

    x, y = a, b
    while y != 0:
        x, y = y, x % y
    gcd_value = x

    lcm_value = a * b // gcd_value

    if combo.get() == 'GCD': 
        result_label.config(text= f'GCD(m, n) = {gcd_value}')
    else: 
        result_label.config(text= f'LCM(m, n) = {lcm_value}')

tk.Button(root, 
          text= 'Calculate', 
          command= calculate).grid(row= 4, 
                                   column= 0, 
                                   columnspan= 2, 
                                   pady= 10)

root.mainloop()
```

```python
#cau 11
root = tk.Tk()
root.title('Positive to Negative')
root.geometry('350x180')

tk.Label(root, text= 'Enter p√≥itive year:').grid(row=0, 
                                                 column=0, 
                                                 padx=10, 
                                                 pady=10, 
                                                 sticky='e')
entry_year = tk.Entry(root)
entry_year.grid(row=0, 
                column=1, 
                padx=10, 
                pady=10)

result_label = tk.Label(root, 
                        text= ' ', 
                        font=('Arial', 10))
result_label.grid(row=2, 
                  column=0, 
                  columnspan=2, 
                  pady=20)

def convert():
    try:
        year = int(entry_year.get())
        if year <= 0:
            result_label.config(text= 'Please enter th freaking positive year')
            return

        cycle_number = (year - 4) % 60 + 1
        result_label.config(text=str(cycle_number))
    except ValueError:
        result_label.config(text= "do not redeem the card")

tk.Button(root, text= 'Convert', 
          command=convert).grid(row=3, 
                                column=0, 
                                columnspan=2, 
                                pady=10)

root.mainloop()
```

```python
#cau 12
root = tk.Tk()
root.title('BMI')
root.geometry('400x250')

tk.Label(root, 
         text= 'Enter weight (kg):').grid(row=0, 
                                         column=0, 
                                         padx=10, 
                                         pady=10, 
                                         sticky='e')
entry_weight = tk.Entry(root)
entry_weight.grid(row=0, 
                  column=1, 
                  padx=10, 
                  pady=10)

tk.Label(root, text= 'Enter height (m):').grid(row=1, 
                                               column=0, 
                                               padx=10, 
                                               pady=10, 
                                               sticky='e')
entry_height = tk.Entry(root)
entry_height.grid(row=1, 
                  column=1, 
                  padx=10, 
                  pady=10)

def calculate():
    try:
        weight = float(entry_weight.get())
        height = float(entry_height.get())
        if weight <= 0 or height <= 0:
            result_label.config(text= 'Please enter positive values!')
            return
        
        bmi = weight / (height ** 2)
        bmi_value = round(bmi, 2)

        if bmi < 18.5:
            conclusion = 'Underweight'
        elif 18.5 <= bmi < 24.9:
            conclusion = 'Normal weight'
        elif 25 <= bmi < 29.9:
            conclusion = 'Overweight'
        else:
            conclusion = 'Obese'

        result_label.config(text=f'BMI: {bmi_value}\nConclusion: {conclusion}')

    except ValueError:
        result_label.config(text= 'Enter the freaking valid numbers!')

tk.Button(root, 
          text= 'BMI', 
          command= calculate).grid(row=2, 
                                   column=0, 
                                   columnspan=2, 
                                   pady=10)

result_label = tk.Label(root, 
                        text= ' ', 
                        font=('Arial', 10))
result_label.grid(row=3, 
                  column=0, 
                  columnspan=2, 
                  pady=10)

root.mainloop()
```

```python
#cau 13
root = tk.Tk()
root.title("Python Homework App")
root.geometry("300x250")

current_hw_window = None

def show_notification(chapter, homework):
    messagebox.showinfo("Homework Notification", f"You clicked Homework {chapter}.{homework}")

def open_homework_window(event, chapter):
    global current_hw_window
    if current_hw_window and current_hw_window.winfo_exists():
        current_hw_window.destroy()

    hw_win = tk.Toplevel(root)
    current_hw_window = hw_win
    hw_win.title(f"Chapter {chapter} Homeworks")
    hw_win.geometry("180x400") 
    hw_win.wm_transient(root)  

    frame = tk.Frame(hw_win, bg="lightyellow", bd=1, relief="solid")
    frame.pack(fill="both", expand=True, padx=5, pady=5)

    tk.Label(frame, text=f"Chapter {chapter} Homeworks", bg="lightyellow",
             font=("Arial", 10, "bold")).pack(pady=5)

    for i in range(1, 10):
        btn = tk.Button(frame, text=f"{chapter}.{i}", width=15, bg="white", fg="black",
                        command=lambda c=chapter, n=i: show_notification(c, n))
        btn.pack(pady=2)

        def on_enter(e, b=btn):
            b.config(bg="blue", fg="white")
        def on_leave(e, b=btn):
            b.config(bg="white", fg="black")

        btn.bind("<Enter>", on_enter)
        btn.bind("<Leave>", on_leave)

    # EXIT button with hover effect red
    exit_btn = tk.Button(frame, text="EXIT", width=15, bg="white", fg="black",
                         command=hw_win.destroy)
    exit_btn.pack(pady=10)

    def exit_enter(e):
        exit_btn.config(bg="red", fg="white")
    def exit_leave(e):
        exit_btn.config(bg="white", fg="black")

    exit_btn.bind("<Enter>", exit_enter)
    exit_btn.bind("<Leave>", exit_leave)

chapter_frame = tk.Frame(root)
chapter_frame.pack(pady=20)

for ch in range(1, 8):
    btn = tk.Button(chapter_frame, text=f"{ch}", width=4, height=1)
    btn.grid(row=0, column=ch-1, padx=2, pady=2)
    btn.bind("<Button-1>", lambda event, c=ch: open_homework_window(event, c))

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Python Homework App")
root.geometry("700x250")
root.configure(bg="#f0f0f0")

# store current popup as attribute of root
root.current_popup = None

def show_notification(chapter, homework):
    messagebox.showinfo("Homework Notification", f"You clicked Homework {chapter}.{homework}")

def show_homework_context(button, chapter):
    # Destroy previous popup if exists
    if hasattr(root, "current_popup") and root.current_popup and root.current_popup.winfo_exists():
        root.current_popup.destroy()

    # Create a popup like context menu
    popup = tk.Toplevel(root)
    root.current_popup = popup
    popup.overrideredirect(True)

    # Position popup at the button
    x = button.winfo_rootx()
    y = button.winfo_rooty() + button.winfo_height()
    popup.geometry(f"+{x}+{y}")

    # Frame to hold buttons
    frame = tk.Frame(popup, bg="lightyellow", bd=1, relief="solid")
    frame.pack()

    # 9 homework buttons with hover effect
    for i in range(1, 10):
        btn = tk.Button(frame, text=f"{chapter}.{i}", width=18, bg="white", fg="black",
                        command=lambda c=chapter, n=i: show_notification(c, n))
        btn.pack(pady=2)
        # Hover glow using bind
        btn.bind("<Enter>", lambda e, b=btn: b.config(bg="blue", fg="white"))
        btn.bind("<Leave>", lambda e, b=btn: b.config(bg="white", fg="black"))

    # EXIT button with hover glow
    exit_btn = tk.Button(frame, text="EXIT", width=18, bg="white", fg="black", command=popup.destroy)
    exit_btn.pack(pady=5)
    exit_btn.bind("<Enter>", lambda e: exit_btn.config(bg="red", fg="white"))
    exit_btn.bind("<Leave>", lambda e: exit_btn.config(bg="white", fg="black"))

# --- Chapter bar ---
chapter_frame = tk.Frame(root, bg="#f0f0f0", height=50)
chapter_frame.pack(fill="x")

chapters = 7
for ch in range(1, chapters + 1):
    btn = tk.Button(chapter_frame, text=f"Chapter {ch}", bg="white", fg="black",
                    bd=1, highlightthickness=0, activebackground="#e0e0e0", activeforeground="black")
    btn.pack(side="left", expand=True, fill="both")
    btn.config(command=lambda b=btn, c=ch: show_homework_context(b, c))

root.mainloop()
```

```python
#cau 14
root = tk.Tk()
root.title('Fake Taxi')
root.geometry('450x300')

tk.Label(root, 
         text= 'Select taxi seat:').grid(row=0, 
                                         column=0, 
                                         padx=10, 
                                         pady=10, 
                                         sticky='e')
seat_var = tk.StringVar()
seat_var.set('4')  
tk.OptionMenu(root, 
              seat_var, 
              '4', 
              '7').grid(row=0, 
                        column=1, 
                        padx=10, 
                        pady=10)

tk.Label(root, 
         text= 'Distance (km):').grid(row=1, 
                                      column=0, 
                                      padx=10, 
                                      pady=10, 
                                      sticky='e')
entry_distance = tk.Entry(root)
entry_distance.grid(row=1, 
                    column=1, 
                    padx=10, 
                    pady=10)

tk.Label(root, text= 'Time (minutes):').grid(row=2, 
                                            column=0, 
                                            padx=10, 
                                            pady=10, 
                                            sticky='e')
entry_time = tk.Entry(root)
entry_time.grid(row=2, 
                column=1, 
                padx=10, 
                pady=10)

result_label = tk.Label(root, 
                        text= ' ', 
                        font=('Arial', 
                              10))
result_label.grid(row=4, 
                  column=0, 
                  columnspan=2, 
                  pady=20)

def calculate():
    try:
        seat = int(seat_var.get())
        distance = float(entry_distance.get())
        time = float(entry_time.get())
        
        if distance <= 0 or time < 0:
            result_label.config(text= 'Enter freaking valid distance and time!')
            return
        
        base_fee = 12
        
        if seat == 4:
            if distance <= 30:
                distance_fee = distance * 15.3
            else:
                distance_fee = 30 * 15.3 + (distance - 30) * 12.1
        else:  
            if distance <= 30:
                distance_fee = distance * 16.1
            else:
                distance_fee = 30 * 16.1 + (distance - 30) * 13.8
        
        time_fee = 0
        if time > 5:
            time_fee = (time - 5) * 0.75
        
        # Total fare
        total_fee = base_fee + distance_fee + time_fee
        total_fee = round(total_fee, 2)
        
        result_label.config(text=f"Total fare: ${total_fee}")
        
    except ValueError:
        result_label.config(text='Please enter valid numbers!')

tk.Button(root, 
          text="Calculate Fare", 
          command=calculate).grid(row=3, 
                                  column=0, 
                                  columnspan=2, 
                                  pady=10)

root.mainloop()
```
